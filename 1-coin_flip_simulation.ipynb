```python
from random import randrange
import matplotlib.pyplot as plt

# Store data for visualization
experiments = [100, 1000, 10000, 100000]
heads_counts = []
tails_counts = []

for experiment in experiments:
    heads = tails = 0
    for i in range(experiment):
        if randrange(2) == 0:
            heads += 1
        else:
            tails += 1
    heads_counts.append(heads)
    tails_counts.append(tails)
    print("experiment:", experiment)
    print("heads =", heads, "  tails =", tails)
    print("the ratio of #heads/#tails is", round(heads / tails, 4))
    print()

# Plotting the bar graph
x = range(len(experiments))  # [0, 1, 2, 3]
bar_width = 0.35

plt.bar([i - bar_width/2 for i in x], heads_counts, width=bar_width, label='Heads', color='skyblue')
plt.bar([i + bar_width/2 for i in x], tails_counts, width=bar_width, label='Tails', color='salmon')

plt.xticks(x, [str(e) for e in experiments])
plt.xlabel('Number of Coin Flips')
plt.ylabel('Count')
plt.title('Heads vs Tails in Coin Toss Experiments')
plt.legend()
plt.grid(True, axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
